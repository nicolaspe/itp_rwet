## CMU Pronouncing Dictionary

List of how you pronounce each word. Allison made a library that let's you parse and use it!

```python
import pronouncing
from collections import Counter
```

```python
pronouncing.phones_for_word("abortion")
```

```python
# if we want to know the most common sounds!
text = "april is the cruelest month breeding lilacs out of the dead"
wrds = text.split()
all_phns = [] #phones
```

```python
for item in wrds:
    phns = pronouncing.phones_for_word(item)[0]
    snds = phns.split()
    all_phns.extend(snds)
```

```python
Counter(all_phns).most_common(10)
```

```python
# if a word doesn't exist (in the database...) it will return an empty list
pronouncing.phones_for_word("fümms")
```

```python
# we can also de a "backwards" search! 
# looking for all the words that have a certain chain of sounds
pronouncing.search("IH1 D AH0 L")
```

```python
# the argument is a regular expression, so we can use all the tricks! :D
pronouncing.search("^S P L")
```

```python
# and if we want to know how many syllables are in a word?
programming_phones = pronouncing.phones_for_word("programming")[0]
pronouncing.syllable_count(programming_phones)
```

```python
# in a long string...
print(text)
eliot_phones = [pronouncing.phones_for_word(item)[0] for item in text.split()]
# check the syllables
eliot_syllables = [pronouncing.syllable_count(item) for item in eliot_phones]
print(eliot_syllables)
# total of syllables!
print( sum(eliot_syllables) )
```

```python
# you can also extract the stresses!!!
condescending = pronouncing.phones_for_word("condescending")[0]
pronouncing.stresses(condescending)
```

```python
# and even SEARCH for stresses!
pronouncing.search_stresses('^1001$')
```

```python
# and... let's RHYME :) :) :) 
pronouncing.rhymes("frustration")
```
