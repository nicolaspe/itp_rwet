## [Tracery](http://tracery.io/)

Context-free grammar. Rules for writing text

First, we need to install `tracery`. To use bash commands inside jupyter notebooks, begin the line with `!`

```python
#!pip3 install tracery
```

```python
import tracery
```

```python
rules = {
    "origin" : "Hello, world!"
}
grammar = tracery.Grammar(rules)
```

```python
grammar.flatten("#origin#")
```

```python
rules = {
    "origin": "Hello, #noun#!",
    "noun": "galaxy"
}
grammar = tracery.Grammar(rules)
grammar.flatten("#origin#")
```

Tracery parses according to the created rules, replacing the stuff inside `# ... #`, starting with the rule specified in `tracery.Grammar(<dictionary>).flatten('#rule#')`

```python
rules = {
    "origin": "#greeting#, #noun#! hyd?",
    "noun": "galaxy",
    "greeting": "Howdy"
}
grammar = tracery.Grammar(rules)
grammar.flatten("#origin#")
```

To make tracery pick from various words, we give a `list` to the specific category... Even in the set of rules!!!

```python
rules = {
    "origin": ["#greeting#, #noun#! hyd?", "#greeting#, #people#! SUUUUUP?"],
    "noun": ["galaxy", "world", "neighbourhood", "solar system", "department", "municipality"],
    "people": ["squadmates", "collaborators", "companions", "siblings", "tribes"],
    "greeting": "Howdy"
}
grammar = tracery.Grammar(rules)

for i in range(20):
    print(grammar.flatten("#origin#"))
```

We don't really need to write the lists in the `rules` variable... We can have them on a separate variable!

```python
the_nouns = ["galaxy", "world", "neighbourhood", "solar system", "department", "municipality"]
rules = {
    "origin": ["#greeting#, #noun#! hyd?", "#greeting#, #people#! SUUUUUP?"],
    "noun": the_nouns,
    "people": ["squadmates", "collaborators", "companions", "siblings", "tribes"],
    "greeting": "Howdy"
}
grammar = tracery.Grammar(rules)
grammar.flatten("#origin#")
```

```python
rules = {
    "origin": "#swear.capitalize# #name#! I am #profession.capitalizeAll.a#, not #profession.capitalizeAll.a#!!!",
    "swear": ["dammit", "darn", "alas", "congrats", "eureka", "fiddlesticks", "oops", "thanks"],
    "name": ["Jim", "Don", "Gary", "George", "Larry", "Steven", "Phillip", "Tom", "John"],
    "profession": profession
}
grammar = tracery.Grammar(rules)
grammar.add_modifiers(base_english)
grammar.flatten("#origin#")
```
