Give nlp a text (string) to process

```python
src_txt = open("./sources/emmeline_iincitethismeetingtorebellion.txt").read()
doc = nlp(src_txt)
```

To get a list of the sentences: `doc.sents`

But that's not a string (data type: span), so to use them we need to do: `doc.sents[i].text`

```python
sents = list(doc.sents)
print(type(sents[0]))
print(sents[0].text)
```

```python
# or to make it easier, LIST COMPREHENSION!
sentences = [item.text for item in doc.sents]
print( rng.choice(sentences) )
```

```python
# to get the words...
words = [item.text for item in doc]

# and let's see a few...
for i in range(10):
    print( rng.choice(words) )
```

spacy, also lets us do **[lemmatization](https://nlp.stanford.edu/IR-book/html/htmledition/stemming-and-lemmatization-1.html)**, breaking down the *inflections* of the words

    word     | "lemma"
    ----------------------
    cats     >  cat
    running  >  run
    are      >  be
    geese    >  goose
    
(the pronouns are grouped under `-PRON-`)

Sadly, spacy doesn't do *derivation* (it's waaaay too hard)

    unfixable > un-fix-able

```python
for item in doc:
    print(item.text, item.lemma_)
```

spacy, also let's us do **parts of speech**, in two different ways!
- `item.pos_`
- `item.tag_`: tags is more detailed and granular

To see the specifications, see the [spacy API](https://spacy.io/api/annotation)

```python
for item in doc:
    print(item.text, item.pos_, item.tag_)
```

So, we can extract some types of words!!

```python
[item.text for item in doc if item.tag_ == "NNS"]
```

```python
nouns = [item.text for item in doc if item.pos_ == 'NOUN']
adjectives = [item.text for item in doc if item.pos_ == 'ADJ']

for i in range(10):
    print( rng.choice(adjectives) + ' ' + rng.choice(nouns) )
```

We can also do **noun chunks**

```python
[item.text for item in doc.noun_chunks if len(item.text.split())>2]
```

spacy comes with a full sentence parser/diagrammer (subject and predicate) > **dependency grammar**

Basically, you start with the verb (root of the sentence), and every word is connected to that via a syntactic relation. 
- `item.head.text` : gets you the text it is dependent from
- `item.dep_` : you get the systactical dependency
- `item.children` : the words that are dependent from this word
- `item.subtree` : 

web demo: [displacy](https://demos.explosion.ai/displacy/) > [example](https://demos.explosion.ai/displacy/?text=I%20really%20need%20to%20sleep%20a%20lot%20tonight%2C%20but%20I%20don%27t%20have%20enough%20time&model=en_core_web_sm&cpu=1&cph=1)

```python
target_sentence = rng.choice( list(doc.sents) )

for item in target_sentence:
    print( item.text, item.tag_, item.head.text, item.dep_ , list(item.children) )
```

```python
def flatten_subtree(st):
    return ''.join([w.text_with_ws for w in list(st)]).strip()
```

```python
for item in target_sentence:
    print( item.text, '/', flatten_subtree(item.subtree) )
```

```python
for word in doc:
    if word.dep_ == "prep":
        print( flatten_subtree(word.subtree) )
```
